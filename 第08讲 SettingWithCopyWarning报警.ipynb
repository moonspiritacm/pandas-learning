# 第08讲 SettingWithCopyWarning报警

```python
import pandas as pd
```

## 0. 数据准备

```python
df = pd.read_csv('./datas/beijing_weather/beijing_weather_2018.csv')
df.head(5)
```

```python
df.loc[:, 'bWendu'] = df['bWendu'].str.replace('℃', '').astype('int64')
df.loc[:, 'yWendu'] = df['yWendu'].str.replace('℃', '').astype('int64')
df.head(5)
```

## 1. 问题复现

```python
condition = df["ymd"].str.startswith("2018-03")
condition
```

```python
df[condition]["wen_cha"] = df["bWendu"] - df["yWendu"]
df[condition].head(5)
```

## 2. 解决方案

### 2.1 方案一：一次操作



```python
df.loc[condition, 'wen_cha'] = df['bWendu'] - df['yWendu']
df[condition].head(5)
```

### 2.2 方案二：显式拷贝

```python
df_month3 = df[condition].copy()
df_month3.head(5)
```

```python
df_month3['wen_cha1'] = df['bWendu'] - df['yWendu']
df_month3.head(5)
```
