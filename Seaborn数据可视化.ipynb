# Seaborn数据可视化

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. 外观主题

```python
def sinplot():
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i))
```

### 1.1 matplotlib默认样式

```python
sinplot()
```

### 1.2 darkgrid灰色网格

```python
sns.set_style('darkgrid')
sinplot()
```

### 1.3 whitegrid白色网格

```python
sns.set_style('whitegrid')
sinplot()
```

### 1.4 dark黑色

```python
sns.set_style('dark')
sinplot()
```

### 1.5 white白色

```python
sns.set_style('white')
sinplot()
```

### 1.6 ticks十字叉

```python
sns.set_style('ticks')
sinplot()
```

# 2. 应用实例

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="ticks")

# 准备数据
macro = pd.read_csv('./datas/examples/macrodata.csv')
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
res = np.log(data).diff().dropna()

# 绘制图表
sns.regplot('m1', 'unemp', data=res)

# 添加元素
plt.title('Changes in log %s versus log %s' % ('m1', 'unemp'))

# 图表输出
plt.savefig('Changes in log %s versus log %s.jpg' % ('m1', 'unemp'))
plt.show()
```
